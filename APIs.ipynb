```python
#Getting ped_fatal_df - start by querying people API with these params, then filter down physical conditions
#to exclude impaired by drugs and alcohol
endpoint_people = 'https://data.cityofchicago.org/resource/u6pd-qa9d.json'

params = {
    'person_type':'PEDESTRIAN',
    'injury_classification':'FATAL'
}
res = requests.get(endpoint_people, params = params)
res = res.json()
ped_fatal_df = pd.DataFrame(res)
type(ped_fatal_df)
```

```python
physical_condition_list = ['NORMAL','UNKNOWN','REMOVED BY EMS','OTHER','EMOTIONAL']
ped_fatal_df = ped_fatal_df.loc[ped_fatal_df['physical_condition'].isin(physical_condition_list)]
```
